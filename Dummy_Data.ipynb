```python

import pandas as pd

import random

from faker import Faker



# Initialize Faker

fake = Faker()



# Function to generate random account number

def generate_account_number():

    return fake.bothify(text='#######')



# Function to generate random maturity date in DD/MM/YYYY format

def generate_maturity_date():

    return fake.date_between(start_date='today', end_date='+10y').strftime('%d/%m/%Y')



# Function to generate random float values

def generate_float(min_value, max_value):

    return round(random.uniform(min_value, max_value), 2)



# Function to randomly choose interest type

def generate_interest_type():

    return random.choice(['Fixed', 'Variable'])



# Generate data

data = {

    'First Name': [fake.first_name() for _ in range(200)],

    'Last Name': [fake.last_name() for _ in range(200)],

    'Account Number': [generate_account_number() for _ in range(200)],

    'Maturity Date': [generate_maturity_date() for _ in range(200)],

    'Original Loan Amount': [generate_float(10000, 500000) for _ in range(200)],

    'Current Balance': [],

    'Prepayment': [generate_float(0, 10000) for _ in range(200)],

    'Payment Frequency': ['Monthly' for _ in range(200)],

    'Sub_acc_Current_Balance': [generate_float(1000, 50000) for _ in range(200)],

    'Sub_acc_Interest_Type': [generate_interest_type() for _ in range(200)],

    'Sub_acc_Payment_Frequency': ['Monthly' for _ in range(200)]

}



# Ensure Current Balance is less than Original Loan Amount

for original_loan in data['Original Loan Amount']:

    data['Current Balance'].append(generate_float(1000, original_loan))


# Format numerical fields with thousands separator
df['Original Loan Amount'] = df['Original Loan Amount'].apply(lambda x: f"{x:,.2f}")
df['Current Balance'] = df['Current Balance'].apply(lambda x: f"{x:,.2f}")
df['Prepayment'] = df['Prepayment'].apply(lambda x: f"{x:,.2f}")
df['Sub_acc_Current_Balance'] = df['Sub_acc_Current_Balance'].apply(lambda x: f"{x:,.2f}")


# Create DataFrame

df = pd.DataFrame(data)



df
```
